
# Report: A Computationally-Derived Hypothesis for a Protective Locus in Alzheimer's Disease and Aging

## Executive Summary

This report outlines the evidence for a novel protective mechanism against Alzheimer's Disease (AD), originating from a pleiotropic genetic locus on chromosome 22. Through the integration of sQTL, eQTL, GWAS, and TWAS data—supported by existing literature—we hypothesize that a set of common genetic variants modulates AD risk by regulating a core cellular aging pathway. Specifically, CBY1 is a novel case gene.

---

## 1. Identification of a Pleiotropic Genetic Locus on Chromosome 22 with Protective Effects Against AD and Aging

**MultiGWAS ColocBoost** identified 182 colocalized variants among:

* **AD GWAS**: negative association
* **Telomere\_Length PC1 GWAS**: positive association
* **Image\_Aging3 GWAS**: negative association

These variants are associated with seven genes: **DDX17**, **FAM227A**, **CBY1**, **TOMM22**, **JOSD1**, **SUN2**, and **FAM118A**, after overlapping with brain xQTL loci. Among them, **35 variants** specifically overlap with xQTLs of **DDX17**, **CBY1**, and **SUN2**, as shown in ColocBoost results for Telomere Length PC1 GWAS and xQTL data.

## 2. Ciliogenesis and *CBY1* in AD


The investigation began with the observation of a splicing quantitative trait locus (sQTL) in the **Dorsolateral Prefrontal Cortex (DLPFC)**.

* **Biological Plausibility:**
  1. CBY1 is crucial for cilliegenesis in a neurodevelopmental disease Joubert syndrome.
  2. Primary cillia and cilliary signalling (including Wnt/β-catenin pathway and hedgehog pathway, which CBY1 is related with) play important role in aging and [aging-related brain disorder](https://www.sciencedirect.com/science/article/pii/S0969996121003569)
  3. CBY1 is reported to be involved in a close-related PPI interconection between AD, aging and cillia, reported by [Kalyani B. Karunakaran et al.](https://www.nature.com/articles/s41598-020-72024-4) where Alzheimer disease increases_expression of CBY1 in hippocampus of Alzheimer‚Äôs disease patients; compared with the hippocampus of healthy subjects and aging decreases_expression of CBY1 in hippocampus of healthy humans in middle adulthood; compared with fetal life(<8w). 


* **Statistical Evidence:**
  * 1. **MultiGWAS + xQTL ColocBoost:** 
    * this analysis identified 140 pleiotrophic variants for **CBY1**, among which `chr22:38597462:A:G` is the top variant. 
    * top variant is negatively associated to AD_Wightman_Full_2021(coef = -2.1 × 10⁻⁵, z = -3.96), positively associated with Telomere-pc1(coef = 3.1 × 10⁻⁵, z = 5.40) and negatively associated with image_Aging3 (coef = -2.8 × 10⁻⁵, z = -3.01).
    * the strongest inverse relationships are observed in excitatory neuron expression data from DeJager (coef = -8.6 × 10⁻³, z = -12.77) and Kellis (coef = -4.1 × 10⁻³, z = -10.37) studies, followed by moderate to strong negative associations in ROSMAP DLPFC (coef = -3.9 × 10⁻³, z = -6.60), Kellis oligodendrocyte (coef = -1.9 × 10⁻³, z = -5.03), DeJager inhibitory neuron (coef = -7.5 × 10⁻⁴, z = -6.00), Kellis inhibitory neuron (coef = -3.0 × 10⁻⁵, z = -6.68), and several splicing QTL variants including AC_lf2_sQTL_chr22:38668132:38670766:clu_143018_+:UP:ENSG00000100211 (coef = -4.5 × 10⁻⁴, z = -6.16), AC_lf2_sQTL_chr22:38670989:38673158:clu_143019_+:UP:ENSG00000100211 (coef = -2.8 × 10⁻⁴, z = -6.04), DLPFC_lf2_sQTL_chr22:38670989:38671069:clu_171691_+:PR:ENSG00000100211 (coef = -5.1 × 10⁻⁵, z = -3.30), DLPFC_lf2_sQTL_chr22:38668132:38670766:clu_171690_+:UP:ENSG00000100211 (coef = -3.0 × 10⁻⁶, z = -6.40), along with aging-related imaging data (coef = -2.8 × 10⁻⁵, z = -3.01) and AD Wightman study results (coef = -2.1 × 10⁻⁵, z = -3.96). In contrast, only two associations show positive relationships: telomere-related measures (Telomere-pc1: coef = 3.1 × 10⁻⁵, z = 5.40) and one specific splicing QTL variant AC_lf2_sQTL_chr22:38671188:38673158:clu_143019_+:PR:ENSG00000100211 (coef = 1.4 × 10⁻⁴, z = 3.47).
  * 2. **TWAS:**
    * The eQTL of *CBY1* is significantly **positively associated** with AD risk. The unproductive splicing event `ROSMAP_AC_sQTL_chr22:38670989:38673158:clu_143019_+:UP:ENSG00000100211` in *CBY1* is **positively associated** with AD risk with marginal significance. The productive splicing event `ROSMAP_DLPFC_sQTL_chr22:38671188:38673158:clu_171691_+:PR:ENSG00000100211` in *CBY1* is **significantly negatively associated** with AD risk.   

* **Potential Hypothesis:**
    The conflicting genetic signals for CBY1—where overall expression (eQTL) is risk-associated, while specific splice isoforms (sQTLs) are either protective or pathogenic—can be explained by a multi-phase model of ciliary response that is dictated by both the stage (severity) and the specific nature (Aβ-dominant vs. Aβ+Tau synergy) of the underlying pathology, given the literature's proposal that only Aβ+Tau synergy can lead to the reduction in the length of primary cillia in hippocampal dentate granule cells of 3xAD transgenic mice.


  Phase 1: The Compensatory Elongation Phase (Early-to-Mid Stage AD / Aβ-Dominant Pathology)

  This phase is driven primarily by the accumulation of Aβ1-42, preceding significant tau pathology.


   * Cellular State: In response to initial Aβ-induced stress, the cell initiates a compensatory, but ultimately maladaptive, signaling  response. The literature shows this can manifest as an increase in primary cilia length, mediated by pathways like the 5-HT6 serotonin receptor (as seen in APP/PS1 mice). This elongation is not healthy; it's a sign of dysregulated signaling.
   * CBY1's Role: To build these longer cilia, the cell upregulates the machinery for ciliogenesis. This involves preferentially expressing the functional, protective CBY1 splice isoform. This isoform correctly localizes and performs its function, contributing to the physical elongation of the cilium.
   * Reconciling the Data:
       * This explains the protective signal from your sQTL. In individuals whose pathology is primarily in this phase, having the ability to produce this functional CBY1 isoform is associated  with a cellular state that, while stressed, has not yet collapsed.
       * It aligns perfectly with the findings in APP/PS1 mice, where Aβ pathology leads to longer cilia, and blocking this elongation is beneficial.
   * Literature evidence shows that in hippocampus of AD mice, there is an aberrant elevated expression of [somatostatin receptor 3 (SSTR3)](https://www.scopus.com/pages/publications/85029709635?inward), which in [another paper](https://pmc.ncbi.nlm.nih.gov/articles/PMC7952091/#s4), localizes exclusively to excitatory neuron cilia both in vivo and in culture and regulates excitatory synaptic properties.
   * Cillia is also highly related with autophagy in hippocampus where Primary Cillia–autophagy axis is a gateway to mediate communication between blood-borne factors(osteocalcin) and neurons. Speciffically, osteocalcin levels generally differ between males and females, with males typically having higher levels. This might be one of a possibilities of the significant difference between male and female groups for image_Aging3 endophenotypes and AD(higher prevalence in female).

  Phase 2: The Synergistic Collapse Phase (Mid-to-Late Stage AD / Aβ + Tau Synergy)

  This phase is triggered when significant, synergistic Aβ and Tau pathology overwhelm the cell.


   * Cellular State: The combined toxicity of Aβ and hyperphosphorylated tau is profoundly damaging. It impairs core cellular functions, including the RNA splicing machinery. This "second hit" of tau pathology is what [the literature]((https://www.sciencedirect.com/science/article/pii/S0969996121003569)) suggests is required for cilia shortening and severe neurogenesis defects (as seen in 3xTg-AD mice).
   * CBY1's Role: The "danger" signal persists, so the cell continues to drive high overall transcription of the CBY1 gene, explaining the high eQTL signal. However, the now-defective splicing machinery can no longer produce the correct isoform. It begins to erroneously produce a pathogenic CBY1 splice isoform (`ROSMAP_AC_sQTL_chr22:38670989:38673158:clu_143019_+:UP:ENSG00000100211`). This isoform is likely non-functional or dominant-negative, preventing proper localization and leading to a failure to maintain the ciliary structure and reduced length of cillia.
   * Reconciling the Data:
       * This explains why overall CBY1 expression (eQTL) is a risk factor: it's a marker for this late-stage, dysfunctional response.
       * It explains why the pathogenic sQTL is strongly associated with AD risk: it is the molecular product of the collapse.
       * It aligns perfectly with the findings in 3xTg-AD mice, where the combination of Aβ and tau leads to a reduction in cilia length.

## 3. A Shared Cis-Regulatory Hub: *DDX17, TOMM22, JOSD1, FAM227A*
* DDX17: expression is negative associated with normal Aging, AD and brain atrophy indicated by image_Aging3
* TOMM22: subtle effect in normal aging and image_Aging3, hard to interpret without protein colocalization evidence, since it's a membraine marker of mitochondria.
* JOSD1: specifically significant in Exc, JOSD1 appears to be another DUB that affects mitochondrial homeostasis, as [paper reports](https://pmc.ncbi.nlm.nih.gov/articles/PMC9804494/)
* FAM227A: protein function unknown
* SUN2: crucial for nuclear envelop formation and nuclear membrane integrity, expression positively associated with AD mainly in Exc, but hard to interpret...
