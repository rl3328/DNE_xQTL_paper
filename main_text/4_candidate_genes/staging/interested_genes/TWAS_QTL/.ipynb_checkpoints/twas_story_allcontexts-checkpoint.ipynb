```r
library(ggplot2)
library(dplyr)
library(ggrepel)

# --- Process data ---
plot_data_processed <- plot_data %>%
  ungroup() %>%
  mutate(
    # Extract cell type
    cell_type = case_when(
      grepl("Exc", context) ~ "Exc",
      grepl("Inh", context) ~ "Inh",
      grepl("Mic", context) ~ "Mic",
      grepl("OPC", context) ~ "OPC",
      grepl("Oli", context) ~ "Oli",
      grepl("monocyte", context) ~ "monocyte",
      TRUE ~ "bulk"
    ),
    # Extract QTL type
    qtl_type = case_when(
      grepl("eQTL", context) ~ "eQTL",
      grepl("pQTL", context) ~ "pQTL",
      TRUE ~ "other"
    ),
    # Extract cohort
    cohort = case_when(
      grepl("PCC", context) ~ "PCC",
      grepl("AC", context) ~ "AC",
      grepl("DLPFC", context) ~ "DLPFC",
      grepl("Knight", context) ~ "Knight",
      grepl("MSBB", context) ~ "MSBB",
      grepl("ROSMAP", context) ~ "ROSMAP",
      grepl("DeJager", context) ~ "DeJager",
      grepl("mega", context) ~ "mega",
      TRUE ~ "Unknown"
    ),
    cell_qtl_label = paste0(cell_type, "_", qtl_type)
  )

# Fixed order for x-axis contexts
xqtl_context_order <- plot_data_processed %>%
  select(cell_qtl_label, cell_type, qtl_type) %>%
  distinct() %>%
  arrange(cell_type != "bulk",
          factor(cell_type, levels = c("Exc","Inh","Mic","Oli","OPC","monocyte","bulk")),
          qtl_type) %>%
  pull(cell_qtl_label)

# Create gene-specific cell_qtl_label with ordering
plot_data_processed <- plot_data_processed %>%
  mutate(
    gene_cell_qtl = paste0(gene_name, "_", cell_qtl_label),
    cell_qtl_label = factor(cell_qtl_label, levels = xqtl_context_order)
  ) %>%
  group_by(gene_name, gwas_study) %>%
  mutate(gene_gwas_order = cur_group_id()) %>%
  ungroup() %>%
  arrange(gene_name, gwas_study, cell_qtl_label) %>%
  mutate(x_position = as.numeric(interaction(gene_name, gwas_study, cell_qtl_label, 
                                              drop = TRUE, lex.order = TRUE)))

# Calculate width metrics for flexible panel sizing
gene_panel_widths <- plot_data_processed %>%
  group_by(gene_name) %>%
  summarise(
    n_gwas = n_distinct(gwas_study),
    n_cell_qtl = n_distinct(cell_qtl_label),
    n_combinations = n_distinct(gwas_study, cell_qtl_label),
    .groups = "drop"
  ) %>%
  mutate(
    panel_width = n_combinations * 0.8 + n_gwas * 1.0,
    width_ratio = panel_width / sum(panel_width)
  )

# Calculate GWAS study boundaries for rectangles
gwas_boundaries <- plot_data_processed %>%
  group_by(gene_name, gwas_study) %>%
  summarise(
    xmin = min(x_position) - 0.5,
    xmax = max(x_position) + 0.5,
    .groups = "drop"
  )

# Compute GWAS study label positions - centered in the boundary rectangles
gwas_positions <- gwas_boundaries %>%
  mutate(xpos = (xmin + xmax) / 2) %>%
  left_join(
    plot_data_processed %>% 
      select(gene_name, gwas_study, cohort) %>% 
      distinct(),
    by = c("gene_name", "gwas_study")
  )

# Get max y value for positioning GWAS labels
max_y <- max(-log10(plot_data_processed$twas_pval), na.rm = TRUE)

# ---- Build bar plot with facets ----
p_combined <- ggplot(plot_data_processed,
                     aes(x = x_position, y = neg_log_twas_pval)) +
  # Add GWAS study grouping rectangles
  geom_rect(
    data = gwas_boundaries,
    aes(xmin = xmin, xmax = xmax, ymin = -Inf, ymax = Inf),
    fill = NA, color = "gray30", linewidth = 1,
    inherit.aes = FALSE
  ) +
  # Dot plot with GWAS study labels
  geom_point(aes(color = twas_z_sign), size = 2.5, alpha = 0.8) +
  geom_text_repel(
    aes(label = cohort, color = twas_z_sign),
    size = 2.5,
    fontface = "bold",
    box.padding = 0.35,
    point.padding = 0.3,
    segment.color = "grey50",
    segment.size = 0.3,
    max.overlaps = 30,
    min.segment.length = 0,
    direction = "both",
    seed = 123,
    force = 2,
    force_pull = 0.5,
    show.legend = FALSE
  ) +
  # Threshold line
  geom_hline(yintercept = -log10(2.5e-6), linetype = "dashed", color = "red", linewidth = 1) +
  # Facet by gene
  facet_wrap(~gene_name, scales = "free_x", nrow = 1, strip.position = "top") +
  # Color scheme
  scale_fill_manual(
    values = c("Positive" = "#E74C3C", "Negative" = "#3498DB"),
    name = "TWAS Z-score"
  ) +
  # Pattern scale
  scale_pattern_manual(
    values = setNames(c("stripe", "crosshatch", "circle", "none")[1:length(unique(plot_data_processed$gwas_study))],
                      unique(plot_data_processed$gwas_study)),
    name = "GWAS Study"
  ) +
  scale_x_continuous(
    breaks = plot_data_processed$x_position,
    labels = plot_data_processed$cell_qtl_label
  ) +
  theme_bw(base_size = 14) +
  theme(
    strip.background = element_rect(color = "black", fill = "white", linewidth = 1.2),
    strip.text = element_text(face = "bold", size = 14),
    axis.text.x = element_text(angle = 45, hjust = 1, size = 10, face = "bold"),
    axis.title.x = element_blank(),
    panel.grid.major.x = element_blank(),
    panel.spacing = unit(0.5, "lines"),
    legend.position = "bottom",
    plot.margin = margin(10, 10, 10, 10)
  ) +
  labs(y = bquote(-log[10]~italic(p)), fill = "TWAS Z-score", pattern = "GWAS Study") +
  # Annotate GWAS study names above - now centered
  geom_text(
    data = gwas_positions,
    aes(x = xpos, y = max_y * 1.25, label = gwas_study),
    inherit.aes = FALSE,
    size = 4,
    fontface = "bold",
    vjust = 0
  ) +
  coord_cartesian(clip = "off", ylim = c(0, max_y * 1.35))

print(p_combined)

# Calculate optimal total width based on content
total_width <- max(25, sum(gene_panel_widths$panel_width) + 3)

ggsave("gene_plot_by_gene.png", p_combined,
       width = total_width, height = 9, dpi = 300, bg = "white")

# Print width information
print(gene_panel_widths)
cat("\nTotal plot width:", total_width, "inches\n")
cat("\nWidth breakdown:\n")
for(i in 1:nrow(gene_panel_widths)) {
  cat(sprintf("  %s: %.1f inches (%.1f%% of total)\n", 
              gene_panel_widths$gene_name[i],
              gene_panel_widths$panel_width[i],
              gene_panel_widths$width_ratio[i] * 100))
}
```

```r
library(ggplot2)
library(dplyr)
library(ggrepel)

# --- Process data ---
plot_data_processed <- plot_data %>%
  ungroup() %>%
  mutate(
    # Extract cell type
    cell_type = case_when(
      grepl("Exc", context) ~ "Exc",
      grepl("Inh", context) ~ "Inh",
      grepl("Mic", context) ~ "Mic",
      grepl("OPC", context) ~ "OPC",
      grepl("Oli", context) ~ "Oli",
      grepl("monocyte", context) ~ "monocyte",
      TRUE ~ "bulk"
    ),
    # Extract QTL type
    qtl_type = case_when(
      grepl("eQTL", context) ~ "eQTL",
      grepl("pQTL", context) ~ "pQTL",
      TRUE ~ "other"
    ),
    # Extract cohort
    cohort = case_when(
      grepl("PCC", context) ~ "PCC",
      grepl("AC", context) ~ "AC",
      grepl("DLPFC", context) ~ "DLPFC",
      grepl("Knight", context) ~ "Knight",
      grepl("MSBB", context) ~ "MSBB",
      grepl("ROSMAP", context) ~ "ROSMAP",
      grepl("DeJager", context) ~ "DeJager",
      grepl("mega", context) ~ "mega",
      TRUE ~ "Unknown"
    ),
    cell_qtl_label = paste0(cell_type, "_", qtl_type)
  )

# Fixed order for x-axis contexts
xqtl_context_order <- plot_data_processed %>%
  select(cell_qtl_label, cell_type, qtl_type) %>%
  distinct() %>%
  arrange(cell_type != "bulk",
          factor(cell_type, levels = c("Exc","Inh","Mic","Oli","OPC","monocyte","bulk")),
          qtl_type) %>%
  pull(cell_qtl_label)

# Create gene-specific cell_qtl_label with ordering
plot_data_processed <- plot_data_processed %>%
  mutate(
    gene_cell_qtl = paste0(gene_name, "_", cell_qtl_label),
    cell_qtl_label = factor(cell_qtl_label, levels = xqtl_context_order)
  ) %>%
  group_by(gwas_study, gene_name) %>%
  mutate(gene_order = cur_group_id()) %>%
  ungroup() %>%
  arrange(gwas_study, gene_order, cell_qtl_label) %>%
  mutate(x_position = as.numeric(interaction(gwas_study, gene_name, cell_qtl_label, 
                                              drop = TRUE, lex.order = TRUE)))

# Calculate width metrics for flexible panel sizing
gwas_panel_widths <- plot_data_processed %>%
  group_by(gwas_study) %>%
  summarise(
    n_genes = n_distinct(gene_name),
    n_cell_qtl = n_distinct(cell_qtl_label),
    n_combinations = n_distinct(gene_name, cell_qtl_label),
    .groups = "drop"
  ) %>%
  mutate(
    panel_width = n_combinations * 0.8 + n_genes * 1.0,
    width_ratio = panel_width / sum(panel_width)
  )

# Calculate gene boundaries for rectangles
gene_boundaries <- plot_data_processed %>%
  group_by(gwas_study, gene_name) %>%
  summarise(
    xmin = min(x_position) - 0.5,
    xmax = max(x_position) + 0.5,
    .groups = "drop"
  )

# Compute gene label positions
gene_positions <- plot_data_processed %>%
  group_by(gwas_study, gene_name) %>%
  summarise(
    xpos = mean(x_position),
    .groups = "drop"
  )

# Get max y value for positioning gene labels
max_y <- max(-log10(plot_data_processed$twas_pval), na.rm = TRUE)

# ---- Build dot plot with facets ----
p_combined <- ggplot(plot_data_processed,
                     aes(x = x_position, y = neg_log_twas_pval, color = twas_z_sign)) +
  # Add gene grouping rectangles
  geom_rect(
    data = gene_boundaries,
    aes(xmin = xmin, xmax = xmax, ymin = -Inf, ymax = Inf),
    fill = NA, color = "gray30", linewidth = 1,
    inherit.aes = FALSE
  ) +
  geom_point(size = 2.5, alpha = 0.8) +
  # Add cohort labels with repel to avoid overlap
  geom_text_repel(
    aes(label = cohort),
    size = 2.5,
    fontface = "bold",
    color = "black",
    box.padding = 0.35,
    point.padding = 0.3,
    segment.color = "grey50",
    segment.size = 0.3,
    max.overlaps = 30,
    min.segment.length = 0,
    direction = "both",
    seed = 123,
    force = 2,
    force_pull = 0.5
  ) +
  geom_hline(yintercept = -log10(5e-8), linetype = "dashed", color = "red") +
  # KEPT: facet_wrap with square boxes
  facet_wrap(~gwas_study, scales = "free_x", nrow = 1, strip.position = "top") +
  scale_color_manual(values = c("Positive" = "#E74C3C", "Negative" = "#3498DB")) +
  scale_x_continuous(
    breaks = plot_data_processed$x_position,
    labels = plot_data_processed$cell_qtl_label
  ) +
  theme_bw(base_size = 14) +
  theme(
    # KEPT: square facet boxes
    strip.background = element_rect(color = "black", fill = "white", linewidth = 1.2),
    strip.text = element_text(face = "bold", size = 14),
    axis.text.x = element_text(angle = 45, hjust = 1, size = 10, face = "bold"),
    axis.title.x = element_blank(),
    panel.grid.major.x = element_blank(),
    panel.spacing = unit(0.5, "lines"),  # Space between facets
    legend.position = "bottom",
    plot.margin = margin(10, 10, 10, 10)
  ) +
  labs(y = bquote(-log[10]~italic(p)), color = "TWAS Z-score") +
  # Annotate gene names above
  geom_text(
    data = gene_positions,
    aes(x = xpos, y = max_y * 1.25, label = gene_name),
    inherit.aes = FALSE,
    size = 4,
    fontface = "bold.italic",
    vjust = 0
  ) +
  coord_cartesian(clip = "off", ylim = c(0, max_y * 1.35))

print(p_combined)

# Calculate optimal total width based on content
total_width <- max(25, sum(gwas_panel_widths$panel_width) + 3)

ggsave("gene_plot_by_gwas.png", p_combined,
       width = total_width, height = 9, dpi = 300, bg = "white")

# Print width information
print(gwas_panel_widths)
cat("\nTotal plot width:", total_width, "inches\n")
cat("\nWidth breakdown:\n")
for(i in 1:nrow(gwas_panel_widths)) {
  cat(sprintf("  %s: %.1f inches (%.1f%% of total)\n", 
              gwas_panel_widths$gwas_study[i],
              gwas_panel_widths$panel_width[i],
              gwas_panel_widths$width_ratio[i] * 100))
}
```

```r
Aging_cilia_genes_selected = Aging_cilia_genes |> filter(gene_name %in% c("CBY1", "MAP4", "KIF9", "SETD2", "MON1A"))#, "PLXNB1"
   
Aging_cilia_genes_selected = Aging_cilia_genes_selected |> mutate(gwas_study = ifelse(grepl("Bellenguez|Wightman|Jansen", gwas_study), "AD", gwas_study)) |>
  group_by(gwas_study,context) |> 
  slice_min(twas_pval, n = 1, with_ties = FALSE) |> arrange(gene_name, gwas_study, context)
```

```r
plot_data <- Aging_cilia_genes_selected %>%
    mutate(
      # Calculate -log10(bonferroni p-value)
      neg_log_twas_pval = -log10(twas_pval),
      # Create color factor based on twas_z sign
      twas_z_sign = ifelse(twas_z < 0, "Negative", "Positive"),
      # Clean up context names for better display (optional)
      context = str_replace(context, 
                               "^(.+?_sQTL)[^:]*:.*:(clu_\\d+_\\+):([A-Z]+):.*$", 
                               "\\1:\\2:\\3"),
      gwas_study = ifelse(grepl("Jansen|Bellenguez|Wightman", gwas_study), "AD", gwas_study))
plot_data = plot_data[-nrow(plot_data),]
plot_data |> arrange(gene_name, gwas_study, context)
```

```r
# Modified R code for gene-grouped bar plot
library(ggplot2)
library(ggpattern)

# Create position variable for spacing between genes
# Create position variable for spacing between genes - FIXED VERSION
plot_data_processed <- plot_data %>%
  arrange(gene_name, gwas_study, context) %>%     # Arrange FIRST
  group_by(gene_name) %>%
  mutate(context_within_gene = row_number()) %>%  # Then number the rows
  ungroup()

# Calculate counts for each gene
cby1_count <- sum(plot_data_processed$gene_name == "CBY1")
kif9_count <- sum(plot_data_processed$gene_name == "KIF9")
map4_count <- sum(plot_data_processed$gene_name == "MAP4")
setd2_count <- sum(plot_data_processed$gene_name == "SETD2")
mon1a_count <- sum(plot_data_processed$gene_name == "MON1A")
# Calculate x_pos with spacing between genes
plot_data_processed <- plot_data_processed %>%
  mutate(x_pos = case_when(
    gene_name == "CBY1" ~ context_within_gene,
    gene_name == "MAP4" ~ context_within_gene + cby1_count + 1,
    gene_name == "KIF9" ~ context_within_gene + cby1_count + map4_count + 2,
    gene_name == "SETD2" ~ context_within_gene + cby1_count + map4_count + kif9_count + 3,
    gene_name == "MON1A" ~ context_within_gene + cby1_count + map4_count + kif9_count + mon1a_count + 4,
    TRUE ~ NA_real_  # fallback for any unexpected gene names
  ))

# Create the plot with proper spacing
p_gene_grouped_cilia <- plot_data_processed %>%
  ggplot(aes(x = x_pos, y = neg_log_twas_pval)) +
  
  # Patterned bars with correct pattern mapping
  geom_bar_pattern(
    aes(fill = twas_z_sign, pattern = gwas_study),
    stat = "identity", 
    width = 0.7,
    colour = "black",
    size = 0.5,
    pattern_density = 0.3,
    pattern_spacing = 0.02,
    pattern_colour = "white",
    pattern_fill = "black",
    pattern_alpha = 0.7
  ) +
  
  # Threshold line at y = 0.05
  geom_hline(yintercept = 5.6, 
             linetype = "dashed", 
             color = "#34495E", 
             size = 1.2, 
             alpha = 0.8) +
  
  # Color scheme
  scale_fill_manual(
    values = c("Positive" = "#E64B35", "Negative" = "#4DBBD5"),
    name = "TWAS Z-score",
    labels = c("Negative" = "Negative (↓)", "Positive" = "Positive (↑)")
  ) +
  
  # Pattern scale for different GWAS studies - ensure all study values are included
  scale_pattern_manual(
    values = setNames(c("stripe", "crosshatch", "circle", "none")[1:length(unique(plot_data_processed$gwas_study))],
                      unique(plot_data_processed$gwas_study)),
    name = "GWAS Study"
  ) +
  
  # Custom x-axis with proper spacing and labels
  scale_x_continuous(
    breaks = plot_data_processed$x_pos,
    labels = plot_data_processed$context,
    expand = expansion(mult = c(0.05, 0.05))
  ) +
  
  # Y-axis
  scale_y_continuous(
    expand = expansion(mult = c(0, 0.12)),
    breaks = scales::pretty_breaks(n = 6)
  ) +
  
  # Labels
  labs(
    title = "TWAS Results for Genes Involved in \nmicrotubule-Mediated Aging & AD Pathology",
    x = "Context within Gene",
    y = bquote(-log[10]~(italic(p)*"-value"))
    # caption = "Grouped by gene • Bar colors: TWAS Z-score direction • Patterns: GWAS study • Dashed line: p = 0.05"
  ) +
  
  # Add gene group labels
  annotate("text", 
           x = mean(plot_data_processed$x_pos[plot_data_processed$gene_name == "CBY1"]),
           y = max(plot_data_processed$neg_log_twas_pval) * 1.1,
           label = "CBY1", 
           size = 6, 
           fontface = "bold",
           color = "#2C3E50") +

  annotate("text", 
           x = mean(plot_data_processed$x_pos[plot_data_processed$gene_name == "MAP4"]),
           y = max(plot_data_processed$neg_log_twas_pval) * 1.1,
           label = "MAP4", 
           size = 6, 
           fontface = "bold",
           color = "#2C3E50") +


  annotate("text", 
           x = mean(plot_data_processed$x_pos[plot_data_processed$gene_name == "KIF9"]),
           y = max(plot_data_processed$neg_log_twas_pval) * 1.1,
           label = "KIF9", 
           size = 6, 
           fontface = "bold",
           color = "#2C3E50") +
  
  annotate("text", 
           x = mean(plot_data_processed$x_pos[plot_data_processed$gene_name == "SETD2"]),
           y = max(plot_data_processed$neg_log_twas_pval) * 1.1,
           label = "SETD2", 
           size = 6, 
           fontface = "bold",
           color = "#2C3E50") +

  annotate("text", 
           x = mean(plot_data_processed$x_pos[plot_data_processed$gene_name == "MON1A"]),
           y = max(plot_data_processed$neg_log_twas_pval) * 1.1,
           label = "MON1A", 
           size = 6, 
           fontface = "bold",
           color = "#2C3E50") +

  # Theme with increased plot width
  theme_minimal(base_size = 12) +
  theme(
    plot.title = element_text(size = 24, face = "bold", hjust = 0.5, color = "#2C3E50", margin = margin(b = 20)),
    axis.title.x = element_text(size = 14, face = "bold", color = "#34495E", margin = margin(t = 15)),
    axis.title.y = element_text(size = 14, face = "bold", color = "#34495E", margin = margin(r = 15)),
    axis.text.x = element_text(angle = 45, hjust = 1, size = 10, color = "#2C3E50", margin = margin(t = 5)),
    axis.text.y = element_text(size = 11, color = "#2C3E50"),
    legend.position = "right",
    legend.title = element_text(face = "bold", size = 12, color = "#2C3E50"),
    legend.text = element_text(size = 10, color = "#2C3E50"),
    legend.box = "vertical",
    panel.grid.major.y = element_line(color = "#BDC3C7", size = 0.5, linetype = "dotted"),
    panel.grid.minor = element_blank(),
    panel.grid.major.x = element_blank(),
    panel.background = element_rect(fill = "#FDFEFE", color = NA),
    plot.background = element_rect(fill = "#FDFEFE", color = NA),
    plot.caption = element_text(size = 10, color = "#7F8C8D", hjust = 0.5, margin = margin(t = 15)),
    plot.margin = margin(25, 25, 25, 25)
  )

# Display the plot with increased width
print(p_gene_grouped_cilia)
```
