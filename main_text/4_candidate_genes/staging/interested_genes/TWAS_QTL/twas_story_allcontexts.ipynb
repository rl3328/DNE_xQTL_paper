```r

plot_data <- AD2_story_genes_selected %>%
    mutate(
      # Calculate -log10(bonferroni p-value)
      neg_log_twas_pval = -log10(twas_pval),
      # Create color factor based on twas_z sign
      twas_z_sign = ifelse(twas_z < 0, "Negative", "Positive")
    )

# Calculate threshold line for p = 0.05
# threshold_line <- -log10(0.05)
```

```r
library(ggplot2)
library(dplyr)
library(ggrepel)

# --- Process data ---
plot_data_processed <- plot_data %>%
  ungroup() %>%
  mutate(
    # Extract cell type
    cell_type = case_when(
      grepl("Exc", context) ~ "Exc",
      grepl("Inh", context) ~ "Inh",
      grepl("Mic", context) ~ "Mic",
      grepl("OPC", context) ~ "OPC",
      grepl("Oli", context) ~ "Oli",
      grepl("monocyte", context) ~ "monocyte",
      TRUE ~ "bulk"
    ),
    # Extract QTL type
    qtl_type = case_when(
      grepl("eQTL", context) ~ "eQTL",
      grepl("pQTL", context) ~ "pQTL",
      TRUE ~ "other"
    ),
    # Extract cohort
    cohort = case_when(
      grepl("PCC", context) ~ "PCC",
      grepl("AC", context) ~ "AC",
      grepl("DLPFC", context) ~ "DLPFC",
      grepl("Knight", context) ~ "Knight",
      grepl("MSBB", context) ~ "MSBB",
      grepl("ROSMAP", context) ~ "ROSMAP",
      grepl("DeJager", context) ~ "DeJager",
      grepl("mega", context) ~ "mega",
      TRUE ~ "Unknown"
    ),
    # Simplify gwas_study names
    gwas_study = case_when(
      startsWith(gwas_study, "image_") ~ paste0("i", substr(gwas_study, 7, nchar(gwas_study))),
      gwas_study == "Telomere_pc1" ~ "Telomere\n length",
      TRUE ~ gwas_study
    ),
    cell_qtl_label = paste0(cell_type, "_", qtl_type)
  )

# Fixed order for x-axis contexts
xqtl_context_order <- plot_data_processed %>%
  select(cell_qtl_label, cell_type, qtl_type) %>%
  distinct() %>%
  arrange(cell_type != "bulk",
          factor(cell_type, levels = c("Exc","Inh","Mic","Oli","OPC","monocyte","bulk")),
          qtl_type) %>%
  pull(cell_qtl_label)

# Define custom GWAS study order
gwas_order <- c("AD", "Telomere\n length", "iAD2", "iAging2")

# Apply factors
plot_data_processed <- plot_data_processed %>%
  mutate(
    cell_qtl_label = factor(cell_qtl_label, levels = xqtl_context_order),
    gwas_study = factor(gwas_study, levels = gwas_order)
  ) 

# KEY FIX: Create unique gene_cell_qtl combinations WITHIN each GWAS study
# This ensures same gene + cell_qtl_label always gets the same x_position
plot_data_processed <- plot_data_processed %>%
  group_by(gwas_study) %>%
  mutate(
    gene_cell_qtl = paste0(gene_name, "_", cell_qtl_label)
  ) %>%
  # Create position mapping based on unique combinations (not on all rows)
  group_by(gwas_study) %>%
  mutate(
    x_position = as.numeric(factor(gene_cell_qtl, 
                                   levels = unique(gene_cell_qtl[order(gene_name, cell_qtl_label)])))
  ) %>%
  ungroup()

# Calculate width metrics for panel sizing
gwas_panel_widths <- plot_data_processed %>%
  group_by(gwas_study) %>%
  summarise(
    n_genes = n_distinct(gene_name),
    n_cell_qtl = n_distinct(cell_qtl_label),
    n_combinations = n_distinct(gene_name, cell_qtl_label),
    .groups = "drop"
  ) %>%
  mutate(
    panel_width = n_combinations * 0.35 + n_genes * 0.6,
    width_ratio = panel_width / sum(panel_width)
  )

# Calculate gene boundaries for rectangles
gene_boundaries <- plot_data_processed %>%
  group_by(gwas_study, gene_name) %>%
  summarise(
    xmin = min(x_position) - 0.5,
    xmax = max(x_position) + 0.5,
    .groups = "drop"
  )

# Compute gene label positions
gene_positions <- plot_data_processed %>%
  group_by(gwas_study, gene_name) %>%
  summarise(
    xpos = mean(x_position),
    .groups = "drop"
  )

# Get max y value for positioning gene labels
max_y <- max(-log10(plot_data_processed$twas_pval), na.rm = TRUE)

# Get unique x positions and labels for axis (one per unique combination per GWAS)
axis_labels <- plot_data_processed %>%
  select(gwas_study, x_position, cell_qtl_label, gene_cell_qtl) %>%
  distinct() %>%
  arrange(gwas_study, x_position)

# ---- Build plot with flexible panel widths ----
p_combined <- ggplot(plot_data_processed,
                     aes(x = x_position, y = neg_log_twas_pval, color = twas_z_sign)) +
  # Add gene grouping rectangles
  geom_rect(
    data = gene_boundaries,
    aes(xmin = xmin, xmax = xmax, ymin = -Inf, ymax = Inf),
    fill = NA, color = "gray30", linewidth = 1,
    inherit.aes = FALSE
  ) +
  geom_point(size = 2.5, alpha = 0.8) +
  # Add cohort labels with repel
  geom_text_repel(
    aes(label = cohort),
    size = 2.5,
    fontface = "bold",
    color = "black",
    box.padding = 0.35,
    point.padding = 0.3,
    segment.color = "grey50",
    segment.size = 0.3,
    max.overlaps = 30,
    min.segment.length = 0,
    direction = "both",
    seed = 123,
    force = 2,
    force_pull = 0.5
  ) +
  geom_hline(yintercept = -log10(2.5e-6), linetype = "dashed", color = "red") +
  facet_grid(. ~ gwas_study, scales = "free_x", space = "free_x") +
  scale_color_manual(values = c("Positive" = "#E74C3C", "Negative" = "#3498DB")) +
  scale_x_continuous(
    breaks = function(limits) {
      # Get x positions within current panel's range
      current_positions <- axis_labels$x_position[
        axis_labels$x_position >= limits[1] & 
        axis_labels$x_position <= limits[2]
      ]
      unique(current_positions)
    },
    labels = function(breaks) {
      # Map breaks to labels
      sapply(breaks, function(b) {
        matched <- axis_labels[axis_labels$x_position == b, ]
        if(nrow(matched) > 0) {
          as.character(matched$cell_qtl_label[1])
        } else {
          ""
        }
      })
    }
  ) +
  theme_bw(base_size = 16) +
  theme(
    strip.background = element_rect(color = "black", fill = "white", linewidth = 1.2),
    strip.text = element_text(face = "bold", size = 14),
    axis.text.x = element_text(angle = 45, hjust = 1, size = 14, face = "bold"),
    axis.title.x = element_blank(),
    panel.grid.major.x = element_blank(),
    panel.spacing.x = unit(0.5, "lines"),
    legend.position = "bottom",
    plot.margin = margin(10, 10, 10, 10)
  ) +
  labs(y = bquote(-log[10]~italic(p)), color = "TWAS Z-score") +
  # Annotate gene names above
  geom_text(
    data = gene_positions,
    aes(x = xpos, y = max_y * 1.25, label = gene_name),
    inherit.aes = FALSE,
    size = 4,
    fontface = "bold.italic",
    vjust = 0
  ) +
  coord_cartesian(clip = "off", ylim = c(5, max_y * 1.35))

print(p_combined)

# Calculate optimal total width based on content
scale_factor <- 1.2
total_width <- max(20, sum(gwas_panel_widths$panel_width) * scale_factor + 1.5)

ggsave("AD2_gene_plot_by_gwas.png", p_combined,
       width = total_width, height = 9, dpi = 300, bg = "white")

# Print detailed width information
cat("\n=== GWAS Panel Width Information ===\n")
print(gwas_panel_widths %>% select(gwas_study, n_genes, n_cell_qtl, n_combinations, panel_width))
cat("\nTotal plot width:", round(total_width, 1), "inches\n")
cat("\nWidth breakdown:\n")
for(i in 1:nrow(gwas_panel_widths)) {
  cat(sprintf("  %s: %d genes × %d cell_qtl = %d combinations → %.1f inches (%.1f%%)\n", 
              gwas_panel_widths$gwas_study[i],
              gwas_panel_widths$n_genes[i],
              gwas_panel_widths$n_cell_qtl[i],
              gwas_panel_widths$n_combinations[i],
              gwas_panel_widths$panel_width[i],
              gwas_panel_widths$width_ratio[i] * 100))
}
```

```r
plot_data <- TL1_Aging3_genes %>%
    mutate(
      # Calculate -log10(bonferroni p-value)
      neg_log_twas_pval = -log10(twas_pval),
      # Create color factor based on twas_z sign
      twas_z_sign = ifelse(twas_z < 0, "Negative", "Positive"),
      # Clean up context names for better display (optional)
      context = str_replace(context, 
                               "^(.+?_sQTL)[^:]*:.*:(clu_\\d+_\\+):([A-Z]+):.*$", 
                               "\\1:\\2:\\3"),
      gwas_study = ifelse(grepl("Jansen|Bellenguez|Wightman", gwas_study), "AD", gwas_study))
plot_data = plot_data |> filter(gwas_study != "image_Aging5") |> group_by(gene_name,gwas_study,context) |>  slice_min(twas_pval, n = 1, with_ties = FALSE)
plot_data |> arrange(gene_name, gwas_study, context)
```

```r
library(ggplot2)
library(dplyr)
library(ggrepel)

# --- Process data ---
plot_data_processed <- plot_data %>%
  ungroup() %>%
  mutate(
    # Extract cell type from context
    cell_type = case_when(
      grepl("Exc", context) ~ "Exc",
      grepl("Inh", context) ~ "Inh",
      grepl("Mic", context) ~ "Mic",
      grepl("OPC", context) ~ "OPC",
      grepl("Oli", context) ~ "Oli",
      grepl("monocyte", context) ~ "monocyte",
      TRUE ~ "bulk"
    ),
    # Extract QTL type from context
    qtl_type = case_when(
      grepl("eQTL", context) ~ "eQTL",
      grepl("pQTL", context) ~ "pQTL",
      TRUE ~ "other"
    ),
    # Extract cohort from context
    cohort = case_when(
      grepl("PCC", context) ~ "PCC",
      grepl("AC", context) ~ "AC",
      grepl("DLPFC", context) ~ "DLPFC",
      grepl("Knight", context) ~ "Knight",
      grepl("MSBB", context) ~ "MSBB",
      grepl("ROSMAP", context) ~ "ROSMAP",
      grepl("DeJager", context) ~ "DeJager",
      grepl("mega", context) ~ "mega",
      TRUE ~ "Unknown"
    ),
    gwas_study = case_when(
        startsWith(gwas_study, "image_") ~ paste0("i", substr(gwas_study, 7, nchar(gwas_study))),
        gwas_study == "Telomere_pc1" ~ "Telomere length",
        TRUE ~ gwas_study
    ),
    # Create combined label for cell type and QTL
    cell_qtl_label = paste0(cell_type, "_", qtl_type)
  )

# Fixed order for x-axis contexts
xqtl_context_order <- plot_data_processed %>%
  select(cell_qtl_label, cell_type, qtl_type) %>%
  distinct() %>%
  arrange(cell_type != "bulk",
          factor(cell_type, levels = c("Exc","Inh","Mic","Oli","OPC","monocyte","bulk")),
          qtl_type) %>%
  pull(cell_qtl_label)

# Create x-position within each GWAS-gene combination
plot_data_processed <- plot_data_processed %>%
  mutate(
    gene_cell_qtl = paste0(gene_name, "_", cell_qtl_label),
    cell_qtl_label = factor(cell_qtl_label, levels = xqtl_context_order)
  ) %>%
  group_by(gwas_study, gene_name) %>%
  mutate(gene_order = cur_group_id()) %>%
  ungroup() %>%
  arrange(gwas_study, gene_name, cell_qtl_label) %>%
  mutate(x_position = as.numeric(interaction(gwas_study, gene_name, cell_qtl_label, 
                                              drop = TRUE, lex.order = TRUE)))

# Calculate width metrics for flexible panel sizing
gwas_panel_widths <- plot_data_processed %>%
  group_by(gwas_study) %>%
  summarise(
    n_genes = n_distinct(gene_name),
    n_cell_qtl = n_distinct(cell_qtl_label),
    n_combinations = n_distinct(gene_name, cell_qtl_label),
    .groups = "drop"
  ) %>%
  mutate(
    panel_width = n_combinations * 0.8 + n_genes * 1.0,
    width_ratio = panel_width / sum(panel_width)
  )

# Calculate gene boundaries for rectangles within each GWAS study
gene_boundaries <- plot_data_processed %>%
  group_by(gwas_study, gene_name) %>%
  summarise(
    xmin = min(x_position) - 0.5,
    xmax = max(x_position) + 0.5,
    .groups = "drop"
  )

# Compute gene label positions
gene_positions <- plot_data_processed %>%
  group_by(gwas_study, gene_name) %>%
  summarise(
    xpos = mean(x_position),
    .groups = "drop"
  )

# Get max y value for positioning gene labels
max_y <- max(-log10(plot_data_processed$twas_pval), na.rm = TRUE)

# ---- Build dot plot with facets ----
p_combined <- ggplot(plot_data_processed,
                     aes(x = x_position, y = neg_log_twas_pval, color = twas_z_sign)) +
  # Add gene grouping rectangles
  geom_rect(
    data = gene_boundaries,
    aes(xmin = xmin, xmax = xmax, ymin = -Inf, ymax = Inf),
    fill = NA, color = "gray30", linewidth = 1,
    inherit.aes = FALSE
  ) +
  # Main data points
  geom_point(size = 2.5, alpha = 0.8) +
  # Add cohort labels with repel to avoid overlap
  geom_text_repel(
    aes(label = cohort),
    size = 2.5,
    fontface = "bold",
    color = "black",
    box.padding = 0.35,
    point.padding = 0.3,
    segment.color = "grey50",
    segment.size = 0.3,
    max.overlaps = 40,
    min.segment.length = 0,
    direction = "both",
    seed = 123,
    force = 2,
    force_pull = 0.5
  ) +
  # Significance threshold line (p = 2.5e-6)
  geom_hline(yintercept = -log10(2.5e-6), linetype = "dashed", color = "red", linewidth = 0.8) +
  # Facet by GWAS study with square boxes
  facet_wrap(~gwas_study, scales = "free_x", nrow = 1, strip.position = "top") +
  # Color scheme for Z-score direction
  scale_color_manual(
    values = c("Positive" = "#E64B35", "Negative" = "#4DBBD5"),
    name = "TWAS Z-score",
    labels = c("Negative" = "Negative (↓)", "Positive" = "Positive (↑)")
  ) +
  # X-axis with context labels
  scale_x_continuous(
    breaks = plot_data_processed$x_position,
    labels = plot_data_processed$cell_qtl_label
  ) +
  # Styling
  theme_bw(base_size = 14) +
  theme(
    # Square facet boxes
    strip.background = element_rect(color = "black", fill = "white", linewidth = 1.2),
    strip.text = element_text(face = "bold", size = 14),
    axis.text.x = element_text(angle = 45, hjust = 1, size = 10, face = "bold"),
    axis.title.x = element_blank(),
    panel.grid.major.x = element_blank(),
    panel.spacing = unit(0.5, "lines"),
    legend.position = "bottom",
    plot.margin = margin(10, 10, 10, 10),
    plot.title = element_text(size = 16, face = "bold", hjust = 0.5, margin = margin(b = 15))
  ) +
  labs(
    title = "TWAS Results for Genes Involved in EVs-Mediated Aging & AD Pathology",
    y = bquote(-log[10]~italic(p)),
    color = "TWAS Z-score"
  ) +
  # Annotate gene names above each group
  geom_text(
    data = gene_positions,
    aes(x = xpos, y = max_y * 1.25, label = gene_name),
    inherit.aes = FALSE,
    size = 4,
    fontface = "bold.italic",
    vjust = 0
  ) +
  coord_cartesian(clip = "off", ylim = c(0, max_y * 1.35))

print(p_combined)

# Calculate optimal total width based on content
total_width <- max(25, sum(gwas_panel_widths$panel_width) + 3)

ggsave("Functional_gene_plot_by_gwas_EV.png", p_combined,
       width = total_width, height = 9, dpi = 300, bg = "white")

# Print width information
print(gwas_panel_widths)
cat("\nTotal plot width:", total_width, "inches\n")
cat("\nWidth breakdown:\n")
for(i in 1:nrow(gwas_panel_widths)) {
  cat(sprintf("  %s: %.1f inches (%.1f%% of total)\n", 
              gwas_panel_widths$gwas_study[i],
              gwas_panel_widths$panel_width[i],
              gwas_panel_widths$width_ratio[i] * 100))
}
```
