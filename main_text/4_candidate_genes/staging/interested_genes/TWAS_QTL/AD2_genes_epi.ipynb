```r
# Oligodendrocyte Precursor Cells
Kelis_OPC = fread("~/data/analysis_result/snatac_fsusie/top_loci/context_specific/ROSMAP_snATAC_Kelis_OPC.exported.toploci.bed.gz")

# Excitatory Neurons
Kelis_Exc = fread("~/data/analysis_result/snatac_fsusie/top_loci/context_specific/ROSMAP_snATAC_Kelis_Exc.exported.toploci.bed.gz")

# Astrocytes
Kelis_Ast = fread("~/data/analysis_result/snatac_fsusie/top_loci/context_specific/ROSMAP_snATAC_Kelis_Ast.exported.toploci.bed.gz")

# Oligodendrocytes
Kelis_Oli = fread("~/data/analysis_result/snatac_fsusie/top_loci/context_specific/ROSMAP_snATAC_Kelis_Oli.exported.toploci.bed.gz")

# Microglia
Kelis_Mic = fread("~/data/analysis_result/snatac_fsusie/top_loci/context_specific/ROSMAP_snATAC_Kelis_Mic.exported.toploci.bed.gz")

# Inhibitory Neurons
Kelis_Inh = fread("~/data/analysis_result/snatac_fsusie/top_loci/context_specific/ROSMAP_snATAC_Kelis_Inh.exported.toploci.bed.gz")

Kelis_sc = bind_rows(
  OPC = Kelis_OPC,
  Exc = Kelis_Exc,
  Ast = Kelis_Ast,
  Oli = Kelis_Oli,
  Mic = Kelis_Mic,
  Inh = Kelis_Inh,
  .id = "cell_type"
)
```

```r
Kelis_SEZ = Kelis_sc |> filter(`#chr` == "chr16", start > 25000000) #|> pull(variant_ID)
```

```r
Kelis_SEZ$cell_type
Kelis_SEZ$variant_ID
```

```r
Kelis_sc_chr16 = Kelis_sc |> filter(`#chr` == "chr16")
dim(Kelis_sc_chr16)
```

```r
# Define the range
range_start <- 29871158-2000
range_end <- 29899546

# Initialize empty dataframe
all_results <- data.frame()

for (i in 1:nrow(Kelis_sc_chr16)) {
  epi_mark_effects <- as.numeric(strsplit(Kelis_sc_chr16$epi_mark_effects[i], ";")[[1]])
  epi_mark_positions <- as.numeric(strsplit(Kelis_sc_chr16$epi_mark_positions[i], ";")[[1]])
  epi_mark_names <- strsplit(Kelis_sc_chr16$epi_mark_names[i], ";")[[1]]
  
  position_effect_df <- data.frame(
    Row = i,
    cell_type = Kelis_sc_chr16$cell_type[i],
    variant_ID = Kelis_sc_chr16$variant_ID[i],
    Position = epi_mark_positions,
    Effect = epi_mark_effects,
    epi_mark_names = epi_mark_names,
    In_Range = epi_mark_positions >= range_start & epi_mark_positions <= range_end
  )
  
  all_results <- rbind(all_results, position_effect_df)
}

# Check the results
cat("Total rows in combined table:", nrow(all_results), "\n")
cat("Total positions in range:", sum(all_results$In_Range), "\n")
cat("Unique cell types:", length(unique(all_results$cell_type)), "\n")
cat("Unique variant IDs:", length(unique(all_results$variant_ID)), "\n")

# View the first few rows
head(all_results)

# Optional: Check positions within range
in_range_data <- all_results[all_results$In_Range, ]
cat("Rows with positions in target range:", nrow(in_range_data), "\n")
```
