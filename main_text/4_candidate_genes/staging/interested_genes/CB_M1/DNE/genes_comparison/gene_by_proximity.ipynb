```r
# Efficient Gene Identification for 20,000+ Variants using TxDb in R

# Initialize TxDb
txdb <- TxDb.Hsapiens.UCSC.hg38.knownGene

# Extract all genes once (for efficiency)
all_genes <- genes(txdb)

# Function for batch processing many variants
find_genes_batch <- function(variants_df, chr_col = "chr", pos_col = "pos", 
                            id_col = "variant_id", window_size = 500000) {
  
  cat("Processing", nrow(variants_df), "variants...\n")
  
  # Convert variants to GRanges
  variants_gr <- GRanges(
    seqnames = variants_df[[chr_col]],
    ranges = IRanges(start = variants_df[[pos_col]], 
                    end = variants_df[[pos_col]]),
    variant_id = variants_df[[id_col]]
  )
  
  # Expand variants by window size
  expanded_variants <- variants_gr + window_size
  
  # Find overlaps (vectorized operation - very fast)
  overlaps <- findOverlaps(expanded_variants, all_genes)
  
  # Extract results efficiently
  variant_hits <- queryHits(overlaps)
  gene_hits <- subjectHits(overlaps)
  
  # Create results data frame
  results <- data.frame(
    variant_id = mcols(variants_gr)$variant_id[variant_hits],
    variant_chr = as.character(seqnames(variants_gr)[variant_hits]),
    variant_pos = start(variants_gr)[variant_hits],
    gene_id = mcols(all_genes)$gene_id[gene_hits],
    gene_chr = as.character(seqnames(all_genes)[gene_hits]),
    gene_start = start(all_genes)[gene_hits],
    gene_end = end(all_genes)[gene_hits],
    gene_strand = as.character(strand(all_genes)[gene_hits]),
    stringsAsFactors = FALSE
  )
  
  # Calculate distances
  results$distance_to_gene <- abs(results$variant_pos - results$gene_start)
  
  # Filter by window size (double-check)
  results <- results[results$distance_to_gene <= window_size, ]
  
  cat("Found", nrow(results), "variant-gene pairs\n")
  
  return(results)
}

# Optimized function for very large datasets using data.table
find_genes_batch_fast <- function(variants_df, chr_col = "chr", pos_col = "pos", 
                                 id_col = "variant_id", window_size = 500000) {
  
  # Convert to data.table for speed
  variants_dt <- data.table(variants_df)
  setnames(variants_dt, c(chr_col, pos_col, id_col), c("chr", "pos", "variant_id"))
  
  # Convert genes to data.table
  genes_dt <- data.table(
    gene_id = mcols(all_genes)$gene_id,
    chr = as.character(seqnames(all_genes)),
    start = start(all_genes),
    end = end(all_genes),
    strand = as.character(strand(all_genes))
  )
  
  # Set keys for efficient joining
  setkey(variants_dt, chr)
  setkey(genes_dt, chr)
  
  # Join by chromosome first
  joined <- genes_dt[variants_dt, on = "chr", allow.cartesian = TRUE]
  
  # Calculate distances and filter
  joined[, distance := abs(start - pos)]
  nearby <- joined[distance <= window_size]
  
  # Select relevant columns
  results <- nearby[, .(variant_id, variant_chr = chr, variant_pos = pos,
                       gene_id, gene_chr = chr, gene_start = start, 
                       gene_end = end, gene_strand = strand, 
                       distance_to_gene = distance)]
  
  return(as.data.frame(results))
}

# Modified function to add gene symbols AND ENSG IDs
add_gene_annotations <- function(results_df) {
  
  cat("Adding gene symbols and ENSG IDs...\n")
  
  # Get unique gene IDs
  unique_genes <- unique(results_df$gene_id)
  
  # Map to symbols and ENSG IDs
  gene_annotations <- select(org.Hs.eg.db, 
                            keys = unique_genes,
                            columns = c("SYMBOL", "GENENAME", "ENSEMBL"),
                            keytype = "ENTREZID")
  
  # Rename ENSEMBL column to ENSG_ID for clarity
  names(gene_annotations)[names(gene_annotations) == "ENSEMBL"] <- "ENSG_ID"
  
  # Merge back
  results_with_annotations <- merge(results_df, gene_annotations, 
                                   by.x = "gene_id", by.y = "ENTREZID", 
                                   all.x = TRUE)
  
  # Handle cases where multiple ENSG IDs exist for one gene
  # Keep the first one and add a column indicating if there were multiple
  if (any(duplicated(results_with_annotations[, c("variant_id", "gene_id")]))) {
    cat("Warning: Some genes have multiple ENSG IDs. Taking the first one.\n")
    results_with_annotations <- results_with_annotations[
      !duplicated(results_with_annotations[, c("variant_id", "gene_id")]), ]
  }
  
  return(results_with_annotations)
}

# Enhanced function using EnsDb for direct ENSG mapping
add_gene_annotations_ensdb <- function(results_df) {
  
  cat("Adding gene symbols and ENSG IDs using EnsDb...\n")
  
  # Check if EnsDb package is available
  if (!require("EnsDb.Hsapiens.v86", quietly = TRUE)) {
    cat("EnsDb.Hsapiens.v86 not available, falling back to org.Hs.eg.db method\n")
    return(add_gene_annotations(results_df))
  }
  
  ensdb <- EnsDb.Hsapiens.v86
  
  # Get unique gene IDs
  unique_genes <- unique(results_df$gene_id)
  
  # First map Entrez to ENSG using org.Hs.eg.db
  entrez_to_ensg <- select(org.Hs.eg.db, 
                          keys = unique_genes,
                          columns = c("ENSEMBL"),
                          keytype = "ENTREZID")
  
  # Then get gene symbols and names from EnsDb using ENSG IDs
  ensg_ids <- unique(entrez_to_ensg$ENSEMBL[!is.na(entrez_to_ensg$ENSEMBL)])
  
  if (length(ensg_ids) > 0) {
    ensg_symbols <- select(ensdb,
                          keys = ensg_ids,
                          columns = c("SYMBOL", "GENENAME"),
                          keytype = "GENEID")
    
    # Merge the mappings
    gene_annotations <- merge(entrez_to_ensg, ensg_symbols, 
                             by.x = "ENSEMBL", by.y = "GENEID", 
                             all.x = TRUE)
    
    # Rename columns for clarity
    names(gene_annotations)[names(gene_annotations) == "ENSEMBL"] <- "ENSG_ID"
    
    # Merge back to results
    results_with_annotations <- merge(results_df, gene_annotations, 
                                     by.x = "gene_id", by.y = "ENTREZID", 
                                     all.x = TRUE)
  } else {
    # Fallback to org.Hs.eg.db only
    results_with_annotations <- add_gene_annotations(results_df)
  }
  
  # Handle duplicates (some genes may have multiple ENSG IDs)
  if (any(duplicated(results_with_annotations[, c("variant_id", "gene_id")]))) {
    cat("Warning: Some genes have multiple ENSG IDs. Taking the first one.\n")
    results_with_annotations <- results_with_annotations[
      !duplicated(results_with_annotations[, c("variant_id", "gene_id")]), ]
  }
  
  return(results_with_annotations)
}

# Function to process in chunks (for memory efficiency)
process_variants_chunked <- function(variants_df, chunk_size = 5000, 
                                   window_size = 500000) {
  
  n_variants <- nrow(variants_df)
  n_chunks <- ceiling(n_variants / chunk_size)
  
  cat("Processing", n_variants, "variants in", n_chunks, "chunks\n")
  
  results_list <- list()
  
  for (i in 1:n_chunks) {
    start_idx <- (i - 1) * chunk_size + 1
    end_idx <- min(i * chunk_size, n_variants)
    
    cat("Processing chunk", i, "of", n_chunks, 
        "(variants", start_idx, "to", end_idx, ")\n")
    
    chunk_variants <- variants_df[start_idx:end_idx, ]
    chunk_results <- find_genes_batch_fast(chunk_variants, window_size = window_size)
    
    results_list[[i]] <- chunk_results
  }
  
  # Combine all results
  all_results <- rbindlist(results_list)
  
  return(as.data.frame(all_results))
}

# Example usage with sample data
create_sample_variants <- function(n = 20000) {
  
  set.seed(123)
  
  # Create sample variant data
  variants <- data.frame(
    variant_id = paste0("var_", 1:n),
    chr = paste0("chr", sample(1:22, n, replace = TRUE)),
    pos = sample(1000000:200000000, n, replace = TRUE),
    ref = sample(c("A", "T", "G", "C"), n, replace = TRUE),
    alt = sample(c("A", "T", "G", "C"), n, replace = TRUE),
    stringsAsFactors = FALSE
  )
  
  return(variants)
}

# Performance comparison function
benchmark_methods <- function(n_variants = 1000) {
  
  # Create test data
  test_variants <- create_sample_variants(n_variants)
  
  # Method 1: Standard approach
  cat("Testing standard method...\n")
  time1 <- system.time({
    results1 <- find_genes_batch(test_variants)
  })
  
  # Method 2: Fast approach
  cat("Testing fast method...\n")
  time2 <- system.time({
    results2 <- find_genes_batch_fast(test_variants)
  })
  
  cat("Standard method:", time1[3], "seconds\n")
  cat("Fast method:", time2[3], "seconds\n")
  cat("Speedup:", round(time1[3] / time2[3], 2), "x\n")
  
  return(list(standard = results1, fast = results2))
}

# Main analysis function with ENSG ID support
run_variant_analysis <- function(variants_file = NULL, variants_df = NULL, 
                                window_size = 500000, add_annotations = TRUE,
                                use_ensdb = NULL,  # Auto-detect if NULL
                                output_file = "variant_gene_results.tsv") {
  
  # Auto-detect EnsDb availability if not specified
  if (is.null(use_ensdb)) {
    use_ensdb <- require("EnsDb.Hsapiens.v86", quietly = TRUE)
    if (use_ensdb) {
      cat("EnsDb.Hsapiens.v86 detected - using enhanced ENSG mapping\n")
    }
  }
  
  # Load variants
  if (!is.null(variants_file)) {
    cat("Loading variants from", variants_file, "...\n")
    variants_df <- fread(variants_file)
  } else if (is.null(variants_df)) {
    cat("Creating sample data...\n")
    variants_df <- create_sample_variants(20000)
  }
  
  # Process variants
  if (nrow(variants_df) > 10000) {
    # Use chunked processing for large datasets
    results <- process_variants_chunked(variants_df, chunk_size = 5000, 
                                       window_size = window_size)
  } else {
    # Use fast method for smaller datasets
    results <- find_genes_batch_fast(variants_df, window_size = window_size)
  }
  
  # Add gene annotations (symbols and ENSG IDs) if requested
  if (add_annotations) {
    if (use_ensdb) {
      results <- add_gene_annotations_ensdb(results)
    } else {
      results <- add_gene_annotations(results)
    }
  }
  
  # Reorder columns for better readability
  if (add_annotations) {
    column_order <- c("variant_id", "variant_chr", "variant_pos",
                     "gene_id", "ENSG_ID", "SYMBOL", "GENENAME",
                     "gene_chr", "gene_start", "gene_end", "gene_strand",
                     "distance_to_gene")
    
    # Only include columns that exist
    existing_cols <- intersect(column_order, names(results))
    results <- results[, existing_cols]
  }
  
  # Save results
  fwrite(results, output_file, sep = '\t')
  cat("Results saved to", output_file, "\n")
  
  # Summary statistics
  cat("\nSummary:\n")
  cat("Total variant-gene pairs:", nrow(results), "\n")
  cat("Unique variants with nearby genes:", length(unique(results$variant_id)), "\n")
  cat("Unique genes found:", length(unique(results$gene_id)), "\n")
  
  if (add_annotations) {
    cat("Genes with ENSG IDs:", sum(!is.na(results$ENSG_ID)), "\n")
    cat("Genes with symbols:", sum(!is.na(results$SYMBOL)), "\n")
  }
  
  return(results)
}

# Quick start examples:

# Example 1: Process sample data with ENSG IDs (auto-detects EnsDb)
# results <- run_variant_analysis()

# Example 2: Process your own data with ENSG IDs
# results <- run_variant_analysis(variants_file = "my_variants.csv")

# Example 3: Process existing data frame with ENSG IDs
# my_variants <- data.frame(...)
# results <- run_variant_analysis(variants_df = my_variants)

# Example 4: Different window size with ENSG IDs
# results <- run_variant_analysis(window_size = 1000000)  # 1Mb window

# Example 5: Force use of org.Hs.eg.db method only
# results <- run_variant_analysis(use_ensdb = FALSE)

# Example 6: Skip annotations (faster processing)
# results <- run_variant_analysis(add_annotations = FALSE)

cat("Script loaded successfully!\n")
cat("Use run_variant_analysis() to process your variants.\n")
cat("The output TSV file will now include ENSG_ID column.\n")
cat("EnsDb.Hsapiens.v86 detected - enhanced ENSG mapping will be used by default.\n")
```
